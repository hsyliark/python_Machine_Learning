## input 17, hidden layer 30->12, output 1 // relu/softplus/sigmoid, adam, mean_squared_error, 50, 20
### 0.8511

```python
# need library
import numpy as np
import tensorflow as tf
# warning remove
if type(tf.contrib) != type(tf):
    tf.contrib._warning = None

# 딥러닝 구동에 필요한 케라스 함수
from keras.models import Sequential
from keras.layers import Dense

# 실행할 때마다 같은 결과를 얻기 위함
seed = 0
np.random.seed(seed)
tf.set_random_seed(seed)

# surgery patient data
Data_set = np.loadtxt("../dataset/ThoraricSurgery.csv", delimiter = ',') # np.array
Data_set

X = Data_set[:,0:17] # 환자 기록
Y = Data_set[:,17] # 수술 결과

# 딥러닝 구조 결정
model = Sequential()
# danse : 연결
model.add(Dense(30, input_dim = 17, activation = 'relu')) # 1층
model.add(Dense(12, activation = 'softplus')) # 2층
model.add(Dense(1, activation = 'sigmoid')) # 3층

# 딥러닝 실행
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X, Y, epochs = 50, batch_size = 20)

# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
```
