```python
import numpy as np
fp1 = np.polyfit(data_seoul['주차장합계'], data_seoul['차량합계'].astype(int), 1) 
fp1
```

```python
d1_function = np.poly1d(fp1)
x_val = np.linspace(1000, 22000, 100)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(data_seoul['주차장합계'], data_seoul['차량합계'], s=35)
plt.plot(x_val, d1_function(x_val), ls='dashed', lw=3, color='g')
plt.xlabel('차량')
plt.ylabel('주차장')
plt.grid()
plt.show()
```
